```python
import pandas as pd
from pathlib import Path
```

```python
# Load in file
# Store filepath in a variable
titles_data_csv = Path("titles.csv")
credits_data_csv = Path("credits.csv")
```

```python
titles_data_df= pd.read_csv(titles_data_csv)
credits_data_df = pd.read_csv(credits_data_csv)
```

```python
merged_movie_data_df = pd.merge(titles_data_df, credits_data_df, on = "id", how = "outer")
merged_movie_data_df.head()
```

```python
columns_to_delete= ['description','age_certification','runtime','imdb_id','imdb_votes']
merged_movie_data_df.drop(columns=columns_to_delete)
```
